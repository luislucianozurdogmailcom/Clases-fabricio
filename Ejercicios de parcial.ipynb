# Ejercicio 1

Escribir una función que tome dos parámetros que sean txt y nvoc y que se fije si las nvocales estén dentro de txt

```python


def ejercicio_parcial_1(txt,nvoc):
    '''
    Esta función busca las palabras que cumplen la condición de tener nvoc (por ejemplo 3)
    vocales dentro de cada palabra. El resultado es una lista.
    '''
    
    iterador                 = 0;
    letra_anterior           = '';
    bandera_estoy_en_palabra = False;
    letra_auxiliar           = '';
    letras_abecedario        = 'abcdefghijlkmnñopqrstuvwxyz'
    palabra_auxiliar         = '';
    contador_vocales         = 0;
    lista_resultado          = [];
    
    
    for letra in txt:
        
        # Llevo todas las letras a minúscula.
        letra_auxiliar = letra.lower();
        
        # Cuando el texto comienza con una letra
        if letra_auxiliar in letras_abecedario and iterador == 0:
            bandera_estoy_en_palabra = True;
        
        # Reconocer cuando comienza una palabra
        if letra_auxiliar in letras_abecedario and letra_anterior not in letras_abecedario:
        
            bandera_estoy_en_palabra = True;
        
        # Reconocer cuando termina una palabra
        if letra_auxiliar not in letras_abecedario and letra_anterior in letras_abecedario:
        
            bandera_estoy_en_palabra = False;
        
        # Cuando el texto termina con una letra
        if letra_auxiliar in letras_abecedario and len(txt) - 1 == iterador:
            
            bandera_estoy_en_palabra = False;
            palabra_auxiliar        += letra;
            
        # Cuando estoy recorriendo una palabra hacemos la acción:
        if bandera_estoy_en_palabra:
            
            palabra_auxiliar += letra;
            
            if letra_auxiliar in 'aeiou':
                
                contador_vocales += 1;
    
        else:
            
            if contador_vocales == nvoc and len(palabra_auxiliar) > 0:
                
                lista_resultado.append(palabra_auxiliar);
            
            palabra_auxiliar = '';
            contador_vocales = 0;
        
        letra_anterior = letra_auxiliar;
        iterador      += 1;
    
    return lista_resultado;
```

```python
texto = '_A veCes Mmmhh!!!sentimos que solo_faltariA Una gota En el mar!!!!,';
ejercicio_parcial_1(texto,3)
```

# Ejercicio 2

Función que recibe una lista y devuelve el elemento que mas veces se ha repetido dentro de la propia lista, junto con el número de veces que se repitió. Si hay dos que se repiten la misma cantidad de veces, retornar el primero encontrado.


```python
lista_2 = [86];


def mesRepetido(ls):
    
    iterador              = 0;
    lista_auxiliar        = ls;
    elemento_mas_repetido = 0;
    contador_mas_repetido = 0;
    
    for elemento in ls:
        
        # Contador que se resetea con cada elemento
        contador_auxiliar = 0;
        
        # Contamos cuantas veces está repetido un elemento
        for elemento_auxiliar in lista_auxiliar:
            
            if elemento_auxiliar == elemento:
                
                contador_auxiliar += 1;
        
        # Guardamos los elementos mas repetidos y su cantidad de repeticiones

        if contador_mas_repetido < contador_auxiliar:

            elemento_mas_repetido = elemento;
            contador_mas_repetido = contador_auxiliar;
            
        
        # Borramos el elemento que estuvimos revisando anteriormente

        if elemento in lista_auxiliar:
            lista_auxiliar.remove(elemento);
        
    return elemento_mas_repetido, contador_mas_repetido;

mesRepetido(lista_2)
```

# Ejercicio 3

Función que recibe un string y devuelve la palabra que mas veces se ha repetido dentro de la propia lista, junto con el número de veces que se repitió. Si hay dos que se repiten la misma cantidad de veces, retornar el primero encontrado.

```python
texto_1 = 'Función que recibe un string y devuelve la palabra que mas veces se ha repetido dentro de la propia lista, junto con el número de veces que se repitió. Si hay dos que se repiten la misma cantidad de veces, retornar el primero encontrado.';


def palabraRepetida(txt):
    
    iterador                 = 0;
    texto_auxiliar           = txt;
    letra_anterior           = '';
    palabra_auxiliar         = '';
    letras_abecedario        = 'abcdefghijlkmnñopqrstuvwxyz';
    bandera_estoy_en_palabra = False;
    lista_de_palabras        = [];
    
    for letra in txt:
        
        # Llevamos todo a minuscula
        letra_auxiliar = letra.lower();
        
        # Reconocer cuando comienza el texto con una palabra
        if letra_auxiliar in letras_abecedario and iterador == 0:
            bandera_estoy_en_palabra = True;
            
        # Reconocer cuando comienza una palabra en cualquier parte del texto
        if letra_auxiliar in letras_abecedario and letra_anterior not in letras_abecedario:
            bandera_estoy_en_palabra = True;
            
        # Reconocer cuando termina una palabra en el medio del texto
        if letra_auxiliar not in letras_abecedario and letra_anterior in letras_abecedario:
            bandera_estoy_en_palabra = False;
            
        # Reconocer cuando todo el texto termina con una palabra
        if letra_auxiliar in letras_abecedario and iterador == len(txt) - 1:
            
            palabra_auxiliar += letra;
            bandera_estoy_en_palabra = False;
        
        
        # Cuando estamos en una palabra:
        if bandera_estoy_en_palabra:
            
            palabra_auxiliar += letra;
        
        else:
            
            if len(palabra_auxiliar) > 0:
                
                lista_de_palabras.append(palabra_auxiliar);
                
            palabra_auxiliar = '';
            
        letra_anterior = letra_auxiliar;
        iterador += 1;
        
    # Comienza la busqueda de encontrar los repetidos:
    lista_de_palabras_auxiliar = lista_de_palabras;
    contador_mas_repetido      = 0;
    contador_auxiliar          = 0;
    palabra_mas_repetida       = '';

    for elemento in lista_de_palabras:
        
        contador_auxiliar = 0;
        
        for elemento_auxiliar in lista_de_palabras_auxiliar:
            
            if elemento == elemento_auxiliar:
                
                contador_auxiliar += 1;
                
        
        if elemento in lista_de_palabras_auxiliar:
            lista_de_palabras_auxiliar.remove(elemento);
        
        # Entra solo cuando encuentro el mas repetido hasta el momento
        if contador_mas_repetido < contador_auxiliar:
            
            contador_mas_repetido = contador_auxiliar;
            palabra_mas_repetida  = elemento;
    
    return palabra_mas_repetida, contador_mas_repetido;
    
palabraRepetida(texto_1)
```

# Ejercicio 4

Hacer una función booleana llamada es_monovocalica que recibe un string, que tiene una sola palabra. La idea es devolver true cuando esa palabra solo utiliza una vocal, caso contrario devolver False.

Hacer una función crear_lista, que recibe como parámetro un texto y retorna una lista de las palabras monovocálicas y sin repetir.

```python
def es_monovocalica(palabra):
    '''
    Esta función recibe una palabra y retorna true si la palabra solo contiene una clase de vocal.
    '''
    contador_vocal = 0;
    
    # Sumamos 1 por cada vocal DISTINTA encontrada.
    if 'a' in palabra:
        contador_vocal += 1;
        
    if 'e' in palabra:
        contador_vocal += 1;
        
    if 'i' in palabra:
        contador_vocal += 1;
        
    if 'o' in palabra:
        contador_vocal += 1;
        
    if 'u' in palabra:
        contador_vocal += 1;
    
    if contador_vocal == 1:
        return True;
    else:
        return False;

def crear_lista(texto):
    
    letras_abecedario        = 'abcdefghijlkmnñopqrstuvwxyz';
    variable_iteradora       = 0;
    bandera_estoy_en_palabra = False;
    letra_anterior           = '';
    palabra_auxiliar         = '';
    lista_palabras           = [];
    
    for letra in texto:
        
        # Reconocer cuando comienza el texto con una palabra
        if letra in letras_abecedario and variable_iteradora == 0:
            bandera_estoy_en_palabra = True;
            
        # Reconocer cuando comienza una palabra en cualquier parte del texto
        if letra in letras_abecedario and letra_anterior not in letras_abecedario:
            bandera_estoy_en_palabra = True;
            
        # Reconocer cuando termina una palabra en el medio del texto
        if letra not in letras_abecedario and letra_anterior in letras_abecedario:
            bandera_estoy_en_palabra = False;
            
        # Reconocer cuando todo el texto termina con una palabra
        if letra in letras_abecedario and variable_iteradora == len(texto) - 1:
            
            palabra_auxiliar += letra;
            bandera_estoy_en_palabra = False;
        
        
        if bandera_estoy_en_palabra:
            
            palabra_auxiliar += letra;
        
        else:
            
            # La palabra no tiene que ser vacia
            if len(palabra_auxiliar) > 0:
                
                # Agregamos las palabras que encontramos y que son monovocálicas a la lista de palabras
                if es_monovocalica(palabra_auxiliar):
                    
                    lista_palabras.append(palabra_auxiliar);
                
            palabra_auxiliar = '';
            
        variable_iteradora += 1;
        letra_anterior      = letra;
    
    # Vamos a quitar todas las palabras repetidas:
    lista_palabras_sin_repetir = [];
    
    for palabra in lista_palabras:
        
        # Si la palabra no está en la lista sin repetir, la agregamos
        if palabra not in lista_palabras_sin_repetir:
            
            lista_palabras_sin_repetir.append(palabra);
    
    # retornamos la lista pedida por el ejercicio
    return lista_palabras_sin_repetir
    
```
