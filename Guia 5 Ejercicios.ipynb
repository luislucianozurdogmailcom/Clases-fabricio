# Ejercicio 9 Guia 5

Escribir un programa que permita ingresar un texto de longitud indefinida e informe por
pantalla:
a – La cantidad de palabras que contiene el texto
b – La palabra de mayor longitud
c – La palabra de menor longitud
Para cada uno de los puntos anteriores desarrollar una función que recibe el texto como
parámetro y retorna lo correspondiente al punto. Estas tres funciones deben ser
utilizadas por el programa principal el cual informará los resultados. Ejemplo:


```
Ingrese un texto: Si hay problema serio, la nave vuelve a la base
El texto tiene 10 palabras, la de mayor longitud es “problema” y la de menor
longitud es “a”
```

````python
def finder(texto):
    '''
    Esta funcion se encarga de retornar la longitud de palabras
    la palabra con mayor longitud y la de menor longitud.
    Recibe por paŕametro    un string
    '''
    letra_anterior           = '';
    contador_palabras        = 0; # Cuenta solo las palabras
    variable_iteradora       = 0;
    contador_letras_palabras = 0; # Cuenta la cantidad de letras en cada palabra
    bandera_contar_letras    = False;
    largo_palabra_mas_larga  = 0; # Máximo
    largo_palabra_mas_corta  = 1000; # Mínimo
    palabra_auxiliar         = '';
    palabra_mas_larga        = '';
    palabra_mas_corta        = '';
    
    for letra in texto:
        
        # Condicion empieza una palabra
        if letra_anterior == ' ' and letra != ' ':
            
            contador_palabras += 1;
            bandera_contar_letras = True;
            
        # Condición termina una palabra
        if (letra_anterior != ' ' and letra == ' ') or (variable_iteradora == len(texto)-1) :
            
            bandera_contar_letras    = False;
            
            # Condición para encontrar el mínimo
            if largo_palabra_mas_corta > contador_letras_palabras:
                
                largo_palabra_mas_corta = contador_letras_palabras;
                palabra_mas_corta       = palabra_auxiliar;
            
            # Condición para encontrar el máximo
            if largo_palabra_mas_larga < contador_letras_palabras:
                
                largo_palabra_mas_larga = contador_letras_palabras;
                palabra_mas_larga       = palabra_auxiliar;
            
            # Reseteamos la variable contadora de letras
            contador_letras_palabras = 0;
        
        # Condición primer letra del texto
        if variable_iteradora == 0 and letra != ' ':
        
            contador_palabras    += 1;
            bandera_contar_letras = True;
        
        # Condición para contar las letras en las palabras
        if bandera_contar_letras:
            
            contador_letras_palabras += 1;
            palabra_auxiliar         += letra;
        
        if bandera_contar_letras == False:

            palabra_auxiliar = '';
        
        variable_iteradora += 1;
        letra_anterior      = letra;
    
    print(f'Contador palabras: {contador_palabras}');
    print(f'La palabra mas corta es: "{palabra_mas_corta}" y la mas larga es "{palabra_mas_larga}" de largo');
    
finder("Si hay problema serio, la nave vuelve a la base");
````

retorne verdadero si encuentra al menos una palabra en ​t ​conformada exactamente
con los mismos caracteres que conforman ​p sin importar el orden de los mismos en la

```

```

```

```

```

```

````python
def ejercicio_10(texto, palabra):
    '''
    Función que recibe un texto ​t ​y una palabra ​p 
    como parámetros, y retorne verdadero si encuentra al menos 
    una palabra en ​t ​conformada exactamente con los mismos 
    caracteres que conforman ​p sin importar el orden de los 
    mismos en la palabra. 
    '''
    variable_iteradora        = 0;
    palabra_auxiliar          = '';
    bandera_estoy_en_palabra  = False;
    letra_anterior            = '';
    variable_iteradora_2      = 0;
    
    for letra in texto:
        
        # Condicion empieza una palabra
        if (letra_anterior == ' ' and letra != ' ') or (letra != ' ' and variable_iteradora == 0):
            
            bandera_estoy_en_palabra = True
            
        # Condición termina una palabra
        if (letra_anterior != ' ' and letra == ' ') or (variable_iteradora == len(texto)-1):
            
            bandera_estoy_en_palabra = False;
        
        if bandera_estoy_en_palabra == True:
            
            palabra_auxiliar += letra;
            
        if bandera_estoy_en_palabra == False:
            
            for letra_comparacion in palabra:
                
                if letra_comparacion in palabra_auxiliar:
                    # Reemplaza la letra que encontré por nada
                    palabra_auxiliar.replace(letra_comparacion,'',1);
                else:
                    # Si no la encuentra me salgo del bucle.
                    break;
                
                # Incrementamos el segundo iterador
                variable_iteradora_2 += 1;    
                
                if len(palabra) == variable_iteradora_2:
                    print('Cumple la condición')
            
            palabra_auxiliar = '';
        
        variable_iteradora += 1;
        letra_anterior      = letra;
        
        
ejercicio_10('el lo saca','casa')
````
